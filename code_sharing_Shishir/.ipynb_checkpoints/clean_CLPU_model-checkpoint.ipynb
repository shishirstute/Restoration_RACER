## 1. Define System Parameters

Let
- \(N_{\mathrm{hvac}}\) = number of HVAC units  
- \(P_{\mathrm{rated}}\) = rated power per HVAC (kW)  
- \(\phi\) = diversity factor  
- \(k_{\mathrm{norm}}\) = fraction of HVAC load at restoration  
- \(k_{\mathrm{dur}}\) = proportional constant (hours / (°C·hour_outage))  
- \(T_{\mathrm{amb}}\) = ambient temperature (°C)  
- \(T_{\mathrm{outage}}\) = outage duration (hours)  

We compute:
\[
P_{\mathrm{CLPU,thermal}}
= N_{\mathrm{hvac}}\;P_{\mathrm{rated}}\;\phi\;k_{\mathrm{norm}}
\]
\[
D_{\mathrm{peak}} = k_{\mathrm{dur}}\;T_{\mathrm{amb}}\;T_{\mathrm{outage}}
\quad(\text{in hours})
\]

## 2. Compute CLPU Peak & End Time

- **Peak multiplier**:  
  \[
  P_{\mathrm{peak}} = P_{\mathrm{CLPU,thermal}} + P_{PV}(t_{\mathrm{out}}^-)
  \]
- **Peak end** (minutes after midnight):  
  \[
  t_{\mathrm{end}} = t_{\mathrm{rest}} + 60\,D_{\mathrm{peak}}
  \]

## 3. Net Load & CLPU Time‐Series

Define the piecewise CLPU component:
\[
\mathrm{CLPU}(t) =
\begin{cases}
0, & t < t_{\mathrm{out}} \\[6pt]
0, & t_{\mathrm{out}} \le t < t_{\mathrm{rest}} \\[6pt]
P_{\mathrm{peak}} - P_{PV}(t), & t_{\mathrm{rest}} \le t < t_{\mathrm{end}} \\[6pt]
0, & t \ge t_{\mathrm{end}}
\end{cases}
\]
Then net building load:
\[
L_{\mathrm{net}}(t) = \max\bigl(0,\; L_{\mathrm{forecast}}(t) + \mathrm{CLPU}(t)\bigr)
\]


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

load_df = pd.read_csv("15_min_load_forecasts.csv")
forecasted_load = load_df["forecasted_load"].values

N_hvac = 6    # Number of HVAC units
P_rated = 0.3    # Example: Rated power per HVAC unit (kW)
diversity_factor = 0.6
k_norm = 0.8    # Ratio of actual HVAC load at restoration
k_duration = 0.01    # Proportionality constant for peak duration (units: hours / (degree * hour_outage))
Temp_amb = 25    # Ambient temperature (e.g., in Celsius)

T_outage = 1
outage_start_minute = 12 * 60 + 15
restoration_minute = 13 * 60 + 15

pv_df = pd.read_csv("estimated_power_generation_testing_1_day_15_min.csv")
pv_df.iloc[int(outage_start_minute/15):int((restoration_minute+15)/15),
           pv_df.columns.get_loc("estimated_power_generation")] = 0
pv_mod = pv_df["estimated_power_generation"].values

CLPU_peak_thermal = N_hvac * P_rated * diversity_factor * k_norm
D_peak_hours = k_duration * Temp_amb * T_outage
D_peak_minutes = math.ceil(D_peak_hours * 60)
peak_end_minute = restoration_minute + D_peak_minutes

num_points = len(forecasted_load)
intervals = np.arange(num_points)
minutes = intervals * 15

outage_idx = int(outage_start_minute / 15)
last_pv_before = pv_mod[outage_idx - 1]
peak_multiplier = CLPU_peak_thermal + last_pv_before

clpu_component = np.zeros(num_points)
actual_net_load = np.zeros(num_points)

for i, m in enumerate(minutes):
    base = forecasted_load[i]
    pv = pv_mod[i]
    if m < outage_start_minute:
        clpu = 0
        actual = base
    elif m < restoration_minute:
        clpu = 0
        actual = 0
    elif m < peak_end_minute:
        clpu = peak_multiplier - pv
        actual = base + clpu
    else:
        clpu = 0
        actual = base
    clpu_component[i] = clpu
    actual_net_load[i] = max(0, actual)

outage_start_idx = outage_start_minute // 15
restoration_idx = restoration_minute // 15
peak_end_idx = peak_end_minute // 15

plt.figure(figsize=(12, 8))
plt.plot(intervals, forecasted_load, linestyle=':', label='Forecasted Load (kW)')
plt.plot(intervals, actual_net_load, color='red', label='Actual Net Load (kW)')
plt.plot(intervals, clpu_component, linestyle='--', color='orange', label='CLPU Component (kW)')
plt.plot(intervals, pv_mod, linestyle='-.', color='green', label='PV Generation (kW)')
plt.axvline(outage_start_idx, color='black', linestyle='--', label='Outage Start')
plt.axvline(restoration_idx, color='blue', linestyle='--', label='Restoration')
if peak_end_idx < num_points:
    plt.axvline(peak_end_idx, color='purple', linestyle=':', label='CLPU End')
plt.xlim(0, num_points - 1)
plt.xlabel("Time Step (0 to 95)")
plt.ylabel("Power (kW)")
plt.title("15‑min Load, CLPU Surge & PV Generation")
plt.legend()
plt.grid(True)
plt.show()

df = pd.DataFrame({
    'Interval': intervals,
    'MinuteOfDay': minutes,
    'BaseForecast (kW)': forecasted_load,
    'CLPU (kW)': clpu_component,
    'PV (kW)': pv_mod,
    'NetLoad (kW)': actual_net_load
})

mask = (intervals >= restoration_idx) & (intervals <= peak_end_idx + 2)
print(df.loc[mask].round(2))
```
